### Изучим признаки, которые выбраны для изучения
#### 1:  surgery? - категориальный признак
          1 = Yes, it had surgery
          2 = It was treated without surgery
####  2:  Age -  категориальный признак
          1 = Adult horse
          2 = Young (< 6 months) --> похоже ошибка в данных, т.к. 1 и 9, длаю вывод, что 9 это young
####  4:  rectal temperature -  непрерывная величина 
          - linear
          - in degrees celsius.
          - An elevated temp may occur due to infection.
          - temperature may be reduced when the animal is in late shock
          - normal temp is 37.8
          - this parameter will usually change as the problem progresses
               eg. may start out normal, then become elevated because of
                   the lesion, passing back through the normal range as the
                   horse goes into shock        
#### 5:  pulse -  непрерывная величина 
          - linear
          - the heart rate in beats per minute
          - is a reflection of the heart condition: 30 -40 is normal for adults
          - rare to have a lower than normal rate although athletic horses
            may have a rate of 20-25
          - animals with painful lesions or suffering from circulatory shock
            may have an elevated heart rate
#### 6:  respiratory rate  -  дискретная величина 
          - linear
          - normal rate is 8 to 10
          - usefulness is doubtful due to the great fluctuations
#### 7:  temperature of extremities - категориальный признак
          - a subjective indication of peripheral circulation
          - possible values:
               1 = Normal
               2 = Warm
               3 = Cool
               4 = Cold
          - cool to cold extremities indicate possible shock
          - hot extremities should correlate with an elevated rectal temp.
#### 11: pain - a subjective judgement of the horse's pain level - категориальный признак
          - possible values:
               1 = alert, no pain
               2 = depressed
               3 = intermittent mild pain
               4 = intermittent severe pain
               5 = continuous severe pain
          - should NOT be treated as a ordered or discrete variable!
          - In general, the more painful, the more likely it is to require
            surgery
          - prior treatment of pain may mask the pain level to some extent
####  23: outcome - категориальный признак
          - what eventually happened to the horse?
          - possible values:
               1 = lived
               2 = died
               3 = was euthanized

### Ищем выбросы через межквартильный размах

```python
q1 = df_hw['pulse'].quantile(0.25)
q3 = df_hw['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
df_hw_out = df_hw[df_hw['pulse'].between(lower_bound, upper_bound, inclusive=True)]#.sort_values('pulse')
```

```python
q1 = df_hw[['pulse','respiratory rate','rectal temperature']].quantile(0.25)
q3 = df_hw[['pulse','respiratory rate','rectal temperature']].quantile(0.75)

iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
df_hw_out = df_hw[df_hw['pulse'].between(lower_bound[0], upper_bound[0], inclusive=True)]
df_hw_out = df_hw[df_hw['respiratory rate'].between(lower_bound[1], upper_bound[1], inclusive=True)]
df_hw_out = df_hw[df_hw['rectal temperature'].between(lower_bound[2], upper_bound[2], inclusive=True)]
```

```python
df_hw_out[['pulse','respiratory rate','rectal temperature']].describe()
```

```python
pd.concat([df_hw, df_hw_out]).drop_duplicates(keep=False)
```

```python
# как видно, количество пустых строк сократилось 
df_hw_out.isnull().sum()
```

Для начала выполним заполнение категориальных признаков, для pain и outcome, temperature of extremities используем моду 

```python
df_hw_out['pain'].fillna(df_hw_out['pain'].mode()[0], inplace=True)
df_hw_out['outcome'].fillna(df_hw_out['outcome'].mode()[0], inplace=True)
df_hw_out['temperature of extremities'].fillna(df_hw_out['temperature of extremities'].mode()[0], inplace=True)
```

```python
df_hw_out.isnull().sum()
```

Посмотрим, как произвести замену пропусков не категориальных признаков. 
Есть предположение, что пульс будет зависеть от возраста и болевых ощущений

```python
df_hw_out.groupby(['Age','pain'])['pulse'].median()
```

Судя по данным, мое предположение верно, заменим pain медианой в разрезе данных признаков

```python
df_hw_out['pulse'].fillna(df_hw_out.groupby(['Age', 'pain'])['pulse'].transform('median'), inplace=True)
```

Осталось заменить пропуски respiratory rate, т.к. в описании сказано, что полезность сомнительна данного признака, посмотрим, может есть зависимость от возраста. Она наблюдается, воспользуюсь медианой для замены

```python
df_hw_out.groupby(['Age'])['respiratory rate'].median()
```

```python
df_hw_out['respiratory rate'].fillna(df_hw_out.groupby('Age')['respiratory rate'].transform('median'), inplace=True)
```
